```python
from sklearn import preprocessing

for f in features:
    if f.startswith("cat"):
        le = preprocessing.LabelEncoder()
        df_whole.loc[:, f] = le.fit_transform(df_whole.loc[:, f])

print(df_whole[features[3]])
```

```python
def encode(enc, x_train, y_train, x_test):
    x_train_transformed = enc.fit_transform(x_train, y_train)
    x_test_transformed = enc.transform(x_test)
    return x_train_transformed, x_test_transformed
```

```python
df_train = df_whole.loc[df_whole.loc[:, "target"] != -1, :]
df_test = df_whole.loc[df_whole.loc[:, "target"] == -1, :]
print(df_train.shape, df_test.shape)
```

```python
df_whole[cat_features] = df_whole[cat_features].astype('category')
df_whole["target"] = df_whole["target"].astype('category')
```

```python
# do target encoding
from category_encoders import TargetEncoder, WOEEncoder, CountEncoder # TargetEncoder is better on kaggle



te = TargetEncoder(cols=cat_features)
# we = WOEEncoder(cols=cat_features)
# ce = CountEncoder(cols=cat_features)

# xtrain1, xtest1 = encode(te, df_train, df_train.loc[:, "target"], df_test)
# xtrain2, xtest2 = encode(we, df_train.loc[:, cat_features], df_train.loc[:, "target"], df_test.loc[:, cat_features])
# xtrain3, xtest3 = encode(ce, df_train.loc[:, cat_features], df_train.loc[:, "target"], df_test.loc[:, cat_features])

x_train_transformed = te.fit_transform(df_train, df_train.loc[:, "target"])
x_test_transformed = te.transform(df_test)
x_train_tmp = df_train.loc[:, cat_features].join(x_train_transformed, lsuffix="_enc", rsuffix="_mean")
x_test_tmp = df_test.loc[:, cat_features].join(x_test_transformed, lsuffix="_enc", rsuffix="_mean")

# x_train_transformed = xtrain1.join(xtrain2, lsuffix="_enc1", rsuffix="_enc2").join(xtrain3)
# x_test_transformed = xtest1.join(xtest2, lsuffix="_enc1", rsuffix="_enc2").join(xtest3)
# x_train_tmp = df_train.loc[:, cat_features].join(x_train_transformed, lsuffix="", rsuffix="_enc3")
# x_test_tmp = df_test.loc[:, cat_features].join(x_test_transformed, lsuffix="", rsuffix="_enc3")
# print(x_train_transformed.shape, x_test_transformed.shape)
print(x_train_tmp.shape, x_test_tmp.shape)

# target encoding improved by 0.0004 on kaggle
# target encoding > woe
```
